```python
import numpy as np
from matplotlib import pyplot as plt
from matplotlib.ticker import FormatStrFormatter
import pandas as pd
```

```python
from pathlib import Path
```

```python
list(Path().cwd().glob('*xlsx'))
```

```python
df8 =pd.read_csv("9th.csv",header=[0,1],index_col=0)
df8=df8.droplevel(1,axis=1)
df8.index=pd.to_datetime(df8.index)
```

```python
df8.columns
```

```python
plt.figure(num=1, figsize=(15,10))
plt.plot(df8.Base1min,label='Cloud Base Min')
plt.plot(df8.Base1mean, label='Cloud Base Mean')
plt.ylabel('Cloudbase (m)',fontsize=20)
plt.legend(loc='best',fontsize=20)
plt.tick_params(labelsize= 18)
plt.title('9th of November Cloud Base' ,fontsize=20)
plt.xlabel('Time',fontsize=20)
```

```python
df9 =pd.read_csv("9th-data.csv",header=[0,1],index_col=0)
df9=df9.droplevel(1,axis=1)
df9.index=pd.to_datetime(df9.index)
```

```python
df9.columns
```

```python
SW = df9.Sdw - df9.Suw
LW= df9.Ldw- df9.Luw
```

```python
plt.figure(num=1, figsize=(15,10))
plt.plot(SW,"-.",label='Sensible heat Flux',lw=2)
plt.plot(LW,"--",label='Latent heat Flux',lw=2)
plt.plot(df9.G, "-", label ='Ground heat Flux')
plt.legend(loc='best',fontsize=20)
plt.tick_params(labelsize= 18)
plt.ylabel('W m${^-}{^2}$',fontsize=20)
plt.legend(loc='best',fontsize=20)
plt.tick_params(labelsize= 18)
plt.ylabel('W m${^-}{^2}$',fontsize=20)
plt.title('Energy Budget on the 9th',fontsize=20)
plt.xlabel('time (day,Hour)',fontsize=20)
```

```python
plt.figure(num=1, figsize=(20,10))
plt.plot(figsize=(15,200))
plt.plot(df9.Tgrass,marker="<",label='Grass Temp',lw=1,color='green')
plt.plot(df9.Td,marker ='>',label='Dry bulb Temp',lw=1, color='red')
plt.plot(df9.Tw,"-" ,label='Wet bulb temp',lw=1, color='blue')
plt.plot(df9.Tsoil,"--",label ='soil Temp',lw=1,color='brown')
plt.ylabel('Temp $^\circ$C',fontsize=20)
plt.tick_params(labelsize= 17)
plt.legend(loc=4,fontsize=20)
plt.title('9th of November Temperatures',fontsize=20)
plt.xlabel('Time',fontsize=20)
```

```python
plt.figure(num=1, figsize=(15,10))
plt.plot(df9.RH,color='blue')
plt.ylabel('%',fontsize=20)
plt.tick_params(labelsize= 17)
plt.legend(loc=4,fontsize=20)
plt.title('9th of November Raltive Humdity',fontsize=20)
plt.xlabel('Time',fontsize=20)
```
