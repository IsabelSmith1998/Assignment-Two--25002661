```python
import numpy as np
from matplotlib import pyplot as plt
from matplotlib.ticker import FormatStrFormatter
```

```python
data5 = np.genfromtxt("UoR_Data (6).csv", delimiter =',', skip_header=2, filling_values=np.nan)
```

```python
Netradiation = data5[:,15]
Td = data5[:,2]
Tgrass = data5[:,3]
WetBulbT = data5[:,4]
RH = data5[:,5]
Ldw = data5[:,13]
Luw =data5[:,14]
Sdw= data5[:,16]
suw= data5[:,17]
time= data5[:,1]
```

```python
SW= (Sdw-suw)
LW= -(Ldw-Luw)
Rn = LW+SW
```

```python
plt.figure(num=1, figsize=(15,10))
plt.plot(SW,"-.",label='Sn',lw=2)
plt.plot(LW,"--",label='Ln',lw=2)
plt.plot(Netradiation,"-",label='Rn',lw=2)
plt.legend(loc='best',fontsize=20)
plt.tick_params(labelsize= 18)
plt.ylabel('W m${^-}{^2}$',fontsize=20)
plt.title('Energy Budget across a week)',fontsize=20)
plt.xlabel('Number of Points',fontsize=20)
```

```python
plt.figure(num=1, figsize=(15,10))
plt.plot(figsize=(15,200))
plt.plot(Tgrass,marker="<",label='Grass Temp',lw=1,color='green')
plt.plot(Td,marker ='<',label='Dry bulb Temp',lw=1, color='red')
plt.plot(WetBulbT,"-" ,label='Wet bulb temp',lw=1, color='blue')
plt.ylabel('Temp $^\circ$C',fontsize=20)
plt.tick_params(labelsize= 16)
plt.legend(loc=1,fontsize=20)
plt.title('Temp across a general week in October',fontsize=20)
plt.xlabel('Number of Points',fontsize=20)
```
